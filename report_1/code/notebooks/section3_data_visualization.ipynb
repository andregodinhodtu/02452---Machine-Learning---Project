```python
# NOTE:
# Boxplots highlight statistical outliers (values far from the majority, e.g. > 1.5Ã—IQR).
# These may be valid but extreme cases (e.g. age 95).
#
# Plausibility checks flag logically impossible or highly unlikely values (just flags, not removing data here) 
# based on domain knowledge (e.g. negative age, ejection fraction > 100, sodium = 500).

range_rules = {
    "age": (0, 120),
    "ejection_fraction": (1, 100),
    "creatinine_phosphokinase": (0, None),  # should be non-negative
    "platelets": (0, None),                  # should be non-negative
    "serum_creatinine": (0, None),           # should be non-negative
    "serum_sodium": (100, 200),              # conservative physiological interval
    "time": (0, None)                        # should be non-negative
}

def out_of_range_mask(series, low, high):
    mask = pd.Series(False, index=series.index)
    if low is not None:
        mask |= series < low
    if high is not None:
        mask |= series > high
    return mask

range_issues = {}
for col, (lo, hi) in range_rules.items():
    mask = out_of_range_mask(data[col], lo, hi)
    n_bad = int(mask.sum())
    if n_bad > 0:
        range_issues[col] = n_bad
        print(f"[FLAG] {col}: {n_bad} values outside expected range ({lo}, {hi}).")

if not range_issues:
    print("No simple range anomalies detected.")
```

### Outliers count (3.1)

```python
# outliers: IQR & z-score 

# functions to calculate IQR interval and z-score outliers
def iqr_outlier_mask(s: pd.Series, k: float = 1.5) -> pd.Series:
    """
    IQR criterion. -> [Q1 - k*IQR; Q3 + k*IQR]
    """
    s = pd.to_numeric(s, errors="coerce")
    q1, q3 = np.nanpercentile(s, [25, 75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return (s < lower) | (s > upper)

def zscore_outlier_mask(s: pd.Series, thr: float = 3.0) -> pd.Series:
    """
    z-score criterion -> |z| > thr
    """
    s = pd.to_numeric(s, errors="coerce")
    mu = np.nanmean(s)
    sd = np.nanstd(s, ddof=0)
    if sd == 0 or np.isnan(sd):
        return pd.Series(False, index=s.index)
    z = (s - mu) / sd
    return np.abs(z) > thr

def summarize_outliers(df: pd.DataFrame, cols: list, k: float = 1.5, zthr: float = 3.0):
    """
    outlier summary for the numerical features

    For each attribute, computes:
      - Number and percentage of outliers by IQR and z-score methods
      - Descriptive statistics (min, Q1, median, Q3, max, mean, std)
      - Skewness as an indicator of asymmetry in the distribution -> skew
    """
    rows = []
    iqr_masks = {}
    z_masks = {}
    for c in cols:
        s = pd.to_numeric(df[c], errors="coerce")
        iqr_mask = iqr_outlier_mask(s, k=k)
        z_mask   = zscore_outlier_mask(s, thr=zthr)
        iqr_masks[c] = iqr_mask
        z_masks[c]   = z_mask

        n = s.notna().sum()
        iqr_n = int(iqr_mask.sum())
        z_n   = int(z_mask.sum())
        rows.append({
            "attribute": c,
            "n_non_null": n,
            "IQR_outliers": iqr_n,
            "IQR_%": (iqr_n / n * 100) if n else np.nan,
            "Z_outliers": z_n,
            "Z_%": (z_n / n * 100) if n else np.nan,
            "min": np.nanmin(s),
            "q1": np.nanpercentile(s, 25),
            "median": np.nanmedian(s),
            "q3": np.nanpercentile(s, 75),
            "max": np.nanmax(s),
            "mean": np.nanmean(s),
            "std": np.nanstd(s),
            "skew": pd.Series(s).skew()
        })
    summary = pd.DataFrame(rows).sort_values(["IQR_%", "Z_%"], ascending=False)
    return summary, iqr_masks, z_masks

#  selection of numerical attributes 
if "continuous_cols" in globals() and isinstance(continuous_cols, list) and len(continuous_cols) > 0:
    num_cols_for_scan = continuous_cols
else:
    num_cols_for_scan = data.select_dtypes(include=[np.number]).columns.tolist()

outlier_summary, iqr_masks, z_masks = summarize_outliers(data, num_cols_for_scan, k=1.5, zthr=3.0)
display(outlier_summary)  
```

### Distribution of attributes (3.2)

#### Histograms + KDE

```python
# separate numerical and categorical features
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
cat_cols = data.select_dtypes(include=['object']).columns

# Histograms + KDE for numerical features
if len(num_cols) > 0:
    rows = math.ceil(len(num_cols) / 3)
    fig, axes = plt.subplots(rows, 3, figsize=(15, 4*rows))
    axes = axes.flatten()

    for i, col in enumerate(num_cols):
        nunique = data[col].nunique(dropna=True)

        if nunique <= 2:  
            # Binary variable â†’ barplot of frequencies
            sns.countplot(data=data, x=col, hue=col, ax=axes[i], palette="Blues", legend=False)
            axes[i].set_xticks([0, 1])
            axes[i].set_xlabel(col)
            axes[i].set_ylabel("Count")
        elif nunique <= 10:
            # Discrete but not strictly binary
            sns.histplot(data[col], discrete=True, ax=axes[i], color="skyblue")
            axes[i].set_xlabel(col)
        else:
            # Continuous variable â†’ histogram + KDE
            sns.histplot(data[col], kde=True, bins=30, ax=axes[i], color="steelblue")
            axes[i].set_xlabel(col)

        axes[i].set_title(f"{col} distribution")

    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

# Barplots for categorical features 
if len(cat_cols) > 0:
    rows = math.ceil(len(cat_cols) / 3)
    fig, axes = plt.subplots(rows, 3, figsize=(15, 4*rows))
    axes = axes.flatten()

    for i, col in enumerate(cat_cols):
        sns.countplot(x=data[col], ax=axes[i], palette="viridis")
        axes[i].set_title(f"{col} frequency")
        axes[i].tick_params(axis='x', rotation=45)

    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

```python
# Test to confirm the theory that a higher 'time' usually corresponds to target=0 (survival)
# Define cutoff point for low and high time: median of 'time'
median_time = data['time'].median()
print(f"Median time: {median_time}")

# Create column to classify 'time' as low or high
data['time_group'] = data['time'].apply(lambda x: 'Low time' if x <= median_time else 'High time')

# Calculate percentages of target=0 and target=1 in each group
percentages = data.groupby('time_group')['DEATH_EVENT'].value_counts(normalize=True).unstack().fillna(0) * 100

print("\nPercentage distribution of DEATH_EVENT by Time Group:")
print(percentages)

# Result
low_time_death_rate = percentages.loc['Low time', 1]
high_time_survival_rate = percentages.loc['High time', 0]

print(f"\nInterpretation:")
print(f"- In the 'Low time' group, {low_time_death_rate:.2f}% of patients experienced the death event (DEATH_EVENT=1).")
print(f"- In the 'High time' group, {high_time_survival_rate:.2f}% of patients survived (DEATH_EVENT=0).")
```
